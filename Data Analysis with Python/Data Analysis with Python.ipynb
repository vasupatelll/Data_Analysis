```python
# Lets calculate average profit of car sold in 2015 by chevrolet acording to vehicle size
# lets assume that profit per car sold is 10%

car_df["Profit"] = car_df["MSRP"].apply(lambda x:x*10/100)
```

```python
chev["Profit"] = chev["MSRP"].apply(lambda x:x*10/100)
```

```python
# total profit of chevrolet

chev_total_profit = chev["Profit"].sum()
chev_total_profit
```

```python
# total profit generated in the year[2015]
chev_profit_2015=chev[(chev["Year"]==2015)].sum()[15]
chev_profit_2015
```

```python
# total profit generted by chevrolet [Vehicle Size = "Compact"]
chev_profit_2015_compact = chev[(chev["Year"]==2015) & (chev["Vehicle Size"]=="Compact")].sum()[15]
chev_profit_2015_compact
```

```python
# total profit generted by chevrolet [Vehicle Size = "Midsize"]
chev_profit_2015_midsize = chev[(chev["Year"]==2015) & (chev["Vehicle Size"]=="Midsize")].sum()[15]
chev_profit_2015_midsize
```

```python
# total profit generted by chevrolet [Vehicle Size = "Large"]
chev_profit_2015_large = chev[(chev["Year"]==2015) & (chev["Vehicle Size"]=="Large")].sum()[15]
chev_profit_2015_large
```

```python
#  total profit generated by scion [2015]

scion = car_df[(car_df["Make"]=='Scion')]
scion
scion_profit_2015 = scion[(scion["Year"]==2015)].sum()[15]
scion_profit_2015
```

```python
# total profit generted by scion [Vehicle Size = "Compact"]
scion_profit_2015_compact = scion[(scion["Year"]==2015) & (scion["Vehicle Size"]=="Compact")].sum()[15]
```

```python
profit_comparision_2015 = pd.DataFrame({'index':["Total Profit","Total Profit [2015]","Compact","Midsize","Large"],
                                "CHEVROLET" : [chev_total_profit,chev_profit_2015,chev_profit_2015_compact,chev_profit_2015_midsize,chev_profit_2015_large],
                                "SCION": [scion_total_profit,scion_profit_2015,scion_profit_2015_compact,scion_profit_2015_midsize,scion_profit_2015_large]})
profit_comparision_2015.set_index("index",inplace=True)
profit_comparision_2015.rename_axis(index=None,inplace=True)
```

```python
profit_comparision_2015
```
